```python
import pandas as pd
import os
import shutil
```

```python
classes  = os.listdir("train/")
```

```python
valset = []
trset = []
```

```python
for cls in classes:
    path = "train/"
    
    nm = os.listdir(path +"/" + cls )
    count = len(nm)
    os.makedirs
    for fl in range(int(count * 0.2)):
        valset.append(path+cls + "/" + nm[fl])
    for fl in range(int(count * 0.2),count):
        trset.append(path+cls + nm[fl])
        
```
